## Step 3: Data Cleaning

```python
#copy of original data is created to work on so that original data stays untouched
df_clean = df.copy()

# Remove duplicate rows
before = len(df_clean)
df_clean = df_clean.drop_duplicates()
df_clean = df_clean.drop_duplicates(subset=['product_id'])
print(f"Removed {before - len(df_clean)} duplicate rows")
print(f"Remaining rows: {len(df_clean)}")
```

```python
#  Cleaning data by Converting final price by removing currency symbol and commas, then converting to float
df_clean['final_price'] = (
    df_clean['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
df_clean['initial_price'] = pd.to_numeric(df_clean['initial_price'], errors='coerce')

df_clean[['initial_price', 'final_price']].describe()
```

```python
# Handle missing values
# discount is 0 if no discount is offered 
df_clean['discount'] = df_clean['discount'].fillna(0)

# rating is zero for some product that have never been rated 
df_clean['is_rated'] = df_clean['ratings_count'] > 0

#  unknown seller name is filled with 'Unknown'
df_clean['seller_name'] = df_clean['seller_name'].fillna('Unknown')
core_cols = ['product_id', 'title', 'rating', 'ratings_count', 'initial_price',
             'final_price', 'discount', 'category']
df_clean[core_cols].isnull().sum()
```

```python
df_clean['category'] = df_clean['category'].astype('category')
print(f"Final cleaned shape: {df_clean.shape}")
df_clean.dtypes[['rating', 'ratings_count', 'initial_price', 'final_price', 'discount', 'category', 'is_rated']]
```

## Step 4: Feature Engineering

```python
# price_difference calculation
df_clean['price_difference'] = df_clean['initial_price'] - df_clean['final_price']

# discount_percentage calculation
df_clean['discount_pct_calculated'] = (
    (df_clean['price_difference'] / df_clean['initial_price']) * 100
).round(2)

df_clean[['initial_price', 'final_price', 'price_difference', 'discount', 'discount_pct_calculated']].head()
```

```python
# Popularity metric (Rating and Ratings count)
df_clean['popularity_score'] = (
    df_clean['rating'] * np.log1p(df_clean['ratings_count'])
).round(2)

df_clean[['title', 'rating', 'ratings_count', 'popularity_score']].sort_values(
    'popularity_score', ascending=False
).head(10)
```

```python
# Extract Genderwise distribution of product count
def extract_gender(category):
    try:
        crumbs = json.loads(category)
        names = [c.get('name', '') for c in crumbs]
        for g in ['Women', 'Men', 'Unisex', 'Boys', 'Girls', 'Kids']:
            if g in names:
                return g
        return 'Other'
    except (json.JSONDecodeError, TypeError):
        return 'Other'

df_clean['gender'] = df_clean['breadcrumbs'].apply(extract_gender)
df_clean['gender'].value_counts()
```

```python
# Extract brand 
df_clean['brand'] = df_clean['title'].astype(str).str.strip()
df_clean[['brand', 'gender']].head()
```

## Step 7: Insights

### Key Findings

1. **Most products are discounted.** Almost all items show an active discount, with final prices typically well below the initial list prices. This "high initial price, deep discount" model is standard across this platform.

2. **About 11% of products have no ratings.** This is significant because it means these products are new or have low visibility, not that they're low-quality.

3. **A few categories dominate the catalog.** Tops, dresses, shirts, jeans, and sports shoes account for a large portion of all listings. These categories drive most of the business.

4. **Discount amount doesn't strongly predict rating or popularity.** Heavier discounts don't reliably lead to more reviews or higher ratings. Product popularity is mainly driven by the sheer volume of reviews, not by how much it's discounted right now.

5. **Price varies a lot between categories.** Some categories (like accessories) have tight price ranges, while others (like jackets or sarees) show much wider ranges. This matters for inventory planning and pricing decisions.

### What This Means

- Focus on the top categories because most of the revenue comes from them.
- Deep discounting might be necessary for the platform's business model, but it's worth checking margins.
- The products with zero ratings need to promoted in order to increase social presence. 
- Some products are popular without being heavily discounted. These products are good to protect margin on rather than discount further.
